```python
# Import librairies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# temps
t = np.linspace(0, np.pi, 500)

# Bobine
L = 1  # Henri
f = 1  # Hz
phi = 0  # phase
I = 1  # amplitude intensit√©
U = L*I*2*np.pi*f  # amplitude tension
i = I*np.cos(2*np.pi*f*t + phi)
u = U*np.cos(2*np.pi*f*t + phi + np.pi/2)
```

```python
# Plot
fig, ax1 = plt.subplots()

# First y-axis (Current)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel(r"Current (A) $\quad i(t)=I_m\cos \left(\omega t + \phi \right)$", color="tab:blue")
ax1.plot(t, i, color="tab:blue", label="Current (I)")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Second y-axis (Voltage)
ax2 = ax1.twinx()
ax2.set_ylabel(r"Voltage (V) $\quad u(t)=L\omega I_m \cos \left( \omega t + \phi + \dfrac{\pi}{2} \right)$", color="tab:red")
ax2.plot(t, u, color="tab:red", linestyle="dashed", label="Voltage (U)")
ax2.tick_params(axis="y", labelcolor="tab:red")

# draw grid
plt.grid()

# Title
plt.title("Current and Voltage in an Inductor")

# Show the plot
plt.show()
```

Pour une <ins>bobine</ins> on observe que la ***tension*** est **en avance par rapport √† l'intensit√©**.\
Ou encore que l'***intensit√©*** est **en retard de phase par rapport √† la tension**.

La diff√©rence de phase $\quad \phi_{u/i}=\phi_u-\phi_i= +\frac{\pi}{2} \quad$ ou $\quad \phi_{i/u}=\phi_i-\phi_u= -\frac{\pi}{2}$
